```julia
using Covid19, DataFrames, Query, VegaLite, VegaDatasets
```

```julia
states = Covid19.states()
```

```julia
states |> @filter(_.state == "Wisconsin") |> DataFrame
```

```julia
states  |> 
DataFrame |>
@vlplot(
    :line,
    x=:date,
    y=:deaths, 
    width=1200, 
    height=800, 
    color=:state
)
```

```julia
by(states, :state, s -> s.deaths)
```
